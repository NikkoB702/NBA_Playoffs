```python
import pandas as pd
```

```python
csv_file_path = 'C:/Users/Home/NBA_Playoffs/Data/per_100_poss_stats.csv'
df = pd.read_csv(csv_file_path)
df.head()
```

```python
# Show column names and data types
print(df.dtypes)
```

```python
# Show basic statistics for each column
print(df.describe())
```

```python
# Calculate the average points per game
average_points = df['PTS'].mean()
print(f"Average Points per Game: {average_points:.2f}")
```

```python
# Sort teams by points scored, in descending order
sorted_by_points = df.sort_values(by='PTS', ascending=False)
```

```python
sorted_by_points.head()
```

```python
import matplotlib.pyplot as plt
```

```python
# Example: Bar plot of points scored by each team
plt.figure(figsize=(12, 6))
plt.bar(df['Team'], df['PTS'])
plt.xticks(rotation=90)
plt.xlabel('Team')
plt.ylabel('Points')
plt.title('Points Scored by Each Team')
plt.show()
```
