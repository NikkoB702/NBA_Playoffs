```python
import pandas as pd
```

```python
csv_file_path = 'C:/Users/Home/NBA_Playoffs/Data/advanced_stats.csv'
df = pd.read_csv(csv_file_path)
df.head()
```

```python
# Show column names and data types
print(df.dtypes)
```

```python
# Calculate the average pace per game
average_pace = df['Pace'].mean()
print(f"Average Pace per Game: {average_pace:.2f}")
```

```python
# List of playoff teams (replace with the actual team names)
playoff_teams = ["Brooklyn Nets", "Philadelphia 76ers", "Boston Celtics", "New York Knicks", "Cleveland Cavaliers", "Golden State Warriors", "Sacramento Kings", "Milwaukee Bucks", "Denver Nuggets", "Memphis Grizzlies", "Los Angeles Clippers", "Phoenix Suns", "Los Angeles Lakers", "Atlanta Hawks", "TBD", "TBD"]

# Filter the DataFrame to include only playoff teams
playoff_teams_df = df[df['Team'].isin(playoff_teams)]

# Display the filtered DataFrame
print(playoff_teams_df)

# Save the filtered DataFrame to a new CSV file
csv_file_path_playoffs = "playoff_teams.csv"
playoff_teams_df.to_csv(csv_file_path_playoffs, index=False)
```
